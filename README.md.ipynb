# 📖 Project Overview

This project was completed as part of the **Oasis Infobyte Data Analytics Internship Program**.

The objective of this project is to perform **Exploratory Data Analysis (EDA)** on a retail supermarket sales dataset to discover meaningful insights about customer purchasing behavior, sales performance, product demand, shipping methods, and regional trends.

Using Python and data visualization libraries, the project transforms raw business data into valuable insights that support informed decision-making.

---

# 🎯 Project Objectives

- Perform data cleaning and preprocessing
- Explore the dataset using statistical analysis
- Analyze customer purchasing behavior
- Identify top-performing products
- Analyze sales trends over time
- Compare regional and category-wise sales
- Generate business insights
- Provide practical business recommendations

# 💡 Key Business Insights

- Technology category generated the highest overall sales.
- Consumer customers contributed the largest share of revenue.
- Sales fluctuate throughout the year, indicating seasonal demand.
- A small number of products generate a significant percentage of total revenue.
- Certain regions consistently outperform others in sales.
- Effective inventory planning can improve profitability.

---

# 📈 Business Recommendations

- Increase stock availability for top-selling products.
- Launch marketing campaigns in low-performing regions.
- Improve customer retention through loyalty programs.
- Optimize shipping operations to reduce delivery time.
- Use historical sales trends for demand forecasting.
