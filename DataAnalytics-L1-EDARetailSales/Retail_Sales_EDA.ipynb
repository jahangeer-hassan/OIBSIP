#Step 2 (Import Libraries)

```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")
```

#Step 3 (Load Dataset)

```
import pandas as pd

df = pd.read_csv("/content/supermarket.csv")

df.head()
```

#Step 4 – Dataset Overview

```
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nℹDataset Information:")
df.info()
```

### Observation

- The dataset contains retail sales transaction records.
- It includes customer information, product details, shipping information, sales, profit, discount, and quantity.
- These variables will be used to identify sales trends and generate business insights.

#Step 5 – Data Quality Assessment

```
print(" Missing Values")
display(df.isnull().sum())

print("\n Duplicate Records")
print(df.duplicated().sum())
```

### Observation

- The dataset was checked for missing values and duplicate records.
- Missing values, if present, will be handled before analysis.
- Duplicate records can affect analysis accuracy and should be removed if found.

#Step 6 – Data Cleaning

```
# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing Postal Code values (if any)
if "Postal Code" in df.columns:
    df["Postal Code"] = df["Postal Code"].fillna(df["Postal Code"].median())

print("Remaining Missing Values:")
display(df.isnull().sum())
```

### Observation

Data cleaning improves the quality of the dataset by removing duplicate records and handling missing values. This ensures that the analysis is based on reliable and consistent data.

#Step 7 – Descriptive Statistics

```
# Descriptive Statistics
df.describe()
```

## Descriptive Statistics

The `describe()` function provides a statistical summary of all numerical columns in the dataset, including count, mean, standard deviation, minimum, maximum, and quartile values. This helps understand the distribution and variability of the data.

#Step 8 – Mean

```
# Mean of Numerical Columns
df.mean(numeric_only=True)
```

### Observation

The mean represents the average value of each numerical feature and provides an overview of the central tendency of the dataset.

#Step 9 – Median

```
# Median of Numerical Columns
df.median(numeric_only=True)
```

### Observation

The median is the middle value of each numerical feature and is less affected by outliers than the mean.

#Step 10 – Mode

```
# Mode of Numerical Columns
df.mode(numeric_only=True)
```

### Observation

The mode represents the most frequently occurring value in each numerical column.

#Step 11 – Standard Deviation

```
# Standard Deviation
df.std(numeric_only=True)
```

### Observation

Standard deviation measures the spread of the data. A higher value indicates greater variability in the dataset.

#Step 12 – Skewness

```
# Skewness
df.skew(numeric_only=True)
```

### Observation

Skewness indicates whether the distribution is symmetric or skewed. Positive values indicate right-skewed distributions, while negative values indicate left-skewed distributions.

#Step 13 – Kurtosis

```
# Kurtosis
df.kurt(numeric_only=True)
```

### Observation

Kurtosis measures the peakedness of the distribution. High kurtosis indicates more extreme values (outliers) compared to a normal distribution.

#Step 14 – Sales Distribution

```
plt.figure(figsize=(8,5))

sns.histplot(df["Sales"], bins=30, kde=True)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

### Observation

- Most sales transactions are of lower values.
- A few transactions have very high sales, indicating the presence of outliers.
- The sales distribution is positively skewed.

#Step 15 – Category-wise Sales

```
plt.figure(figsize=(8,5))

category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sns.barplot(x=category_sales.index,
            y=category_sales.values)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

### Observation

- Technology generated the highest sales.
- Office Supplies and Furniture contributed a significant portion of total revenue.
- The company should continue investing in high-performing product categories.

#Step 16 – Region-wise Sales

```
plt.figure(figsize=(8,5))

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sns.barplot(x=region_sales.index,
            y=region_sales.values)

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

### Observation

- Some regions outperform others in terms of revenue.
- High-performing regions should be prioritized for expansion.
- Low-performing regions may require targeted marketing strategies.

#Step 17 – Customer Segment Analysis

```
plt.figure(figsize=(8,5))

segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

sns.barplot(x=segment_sales.index,
            y=segment_sales.values)

plt.title("Sales by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")

plt.show()
```

### Observation

- Customer segments contribute differently to total sales.
- Identifying the highest-value segment can help improve marketing campaigns.

#Step 18 – Ship Mode Analysis

```
plt.figure(figsize=(8,5))

ship_sales = df.groupby("Ship Mode")["Sales"].sum().sort_values(ascending=False)

sns.barplot(x=ship_sales.index,
            y=ship_sales.values)

plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales")

plt.xticks(rotation=15)

plt.show()
```

### Observation

- Shipping methods vary in their contribution to total sales.
- Understanding shipping preferences can improve logistics planning and customer satisfaction.

#Step 19 – Top 10 Products by Sales

```
plt.figure(figsize=(12,6))

top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

sns.barplot(
    x=top_products.values,
    y=top_products.index,
    palette="viridis"
)

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")

plt.show()
```

### Observation

- The top 10 products contribute a significant share of total revenue.
- These products should be prioritized for inventory management and promotional campaigns.

#Step 20 – Monthly Sales Trend

```
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")
```

```
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

plt.figure(figsize=(14,5))
monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

### Observation

- Sales fluctuate throughout the year.
- Some months show peak sales, indicating seasonal demand.
- Businesses can use this information for inventory planning.

#Step 21 – Quarterly Sales Trend

```
quarterly_sales = df.groupby(df["Order Date"].dt.to_period("Q"))["Sales"].sum()

plt.figure(figsize=(10,5))

quarterly_sales.plot(marker='o', color='green')

plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")

plt.grid(True)

plt.show()
```

### Observation

- Quarterly analysis provides a broader view of business performance.
- It helps identify long-term sales growth and seasonal business cycles.

#Step 22 – Top 10 States by Sales

```
plt.figure(figsize=(10,6))

top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

sns.barplot(
    x=top_states.values,
    y=top_states.index,
    palette="magma"
)

plt.title("Top 10 States by Sales")
plt.xlabel("Total Sales")
plt.ylabel("State")

plt.show()
```

### Observation

- Some states generate much higher sales than others.
- High-performing states can be targeted for business expansion.

#Step 23 – Correlation Heatmap

```
plt.figure(figsize=(8,6))

corr = df.select_dtypes(include="number").corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- The heatmap shows relationships among numerical variables.
- Strong positive or negative correlations help identify important business factors.
